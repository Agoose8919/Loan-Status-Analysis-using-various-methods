```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('Loan Data cleaned.csv')
```

```python
for i in data.select_dtypes('object'):
    data[i]=data[i].astype('category')
```

```python
data['Term']=data['Term'].astype('category')
data['Years in current job']=data['Years in current job'].astype('category')
data['Tax Liens']=data['Tax Liens'].astype('category')
data['Bankruptcies']=data['Bankruptcies'].astype('category')
data['Number of Credit Problems']=data['Number of Credit Problems'].astype('category')
```

```python
data_dummies=pd.get_dummies(columns=list(data.select_dtypes('category')),data=data,dummy_na=True)
```

```python
#testing with simple imputer
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split as tts
```

```python
data_train,data_valid=tts(data,test_size=0.2,stratify=data['Loan Status'],random_state=365)
```

```python
data_train_cat=data_train.select_dtypes('category')
data_valid_cat=data_valid.select_dtypes('category')
```

```python
#conditional imputing
data_train_cat['Loan Status']=data_train['Loan Status']
```

```python
data_train_cat0=pd.DataFrame(SimpleImputer(strategy='most_frequent').fit_transform(data_train_cat.loc[data_train_cat['Loan Status']==0]),columns=data_train_cat.columns)
data_train_cat1=pd.DataFrame(SimpleImputer(strategy='most_frequent').fit_transform(data_train_cat.loc[data_train_cat['Loan Status']==1]),columns=data_train_cat.columns)
```

```python
data_train_cat_imputed=pd.concat([data_train_cat0,data_train_cat1],axis=0)
```

```python
data_train_cat_imputed
```

```python
data_train_num=data_train.select_dtypes('number')
data_valid_num=data_valid.select_dtypes('number')
```

```python
data_train_num0=pd.DataFrame(SimpleImputer(strategy='median').fit_transform(data_train_num.loc[data_train_num['Loan Status']==0]),columns=data_train_num.columns)
data_train_num1=pd.DataFrame(SimpleImputer(strategy='median').fit_transform(data_train_num.loc[data_train_num['Loan Status']==1]),columns=data_train_num.columns)
```

```python
data_train_num_imputed=pd.concat([data_train_num0,data_train_num1],axis=0)
```

```python
data_simple_imputed0=pd.concat([data_train_num_imputed,data_train_cat_imputed.drop('Loan Status',axis=1)],axis=1)
```

```python
data_train_mean0=pd.DataFrame(SimpleImputer(strategy='mean').fit_transform(data_train_num.loc[data_train_num['Loan Status']==0]),columns=data_train_num.columns)
data_train_mean1=pd.DataFrame(SimpleImputer(strategy='mean').fit_transform(data_train_num.loc[data_train_num['Loan Status']==1]),columns=data_train_num.columns)
```

```python
data_train_mean_imputed=pd.concat([data_train_mean0,data_train_mean1],axis=0)
```

```python
data_simple_imputed1=pd.concat([data_train_mean_imputed,data_train_cat_imputed.drop('Loan Status',axis=1)],axis=1)
```

```python
#conditional mean median imputed data for training
data_simple_imputed0.to_csv('Training data(conditional median).csv',index=False)
data_simple_imputed1.to_csv('Training data(conditional mean).csv',index=False)
```

```python
data_train_num01=pd.DataFrame(SimpleImputer(strategy='median').fit_transform(data_train_num),columns=data_train_num.columns)
data_train_cat01=pd.DataFrame(SimpleImputer(strategy='most_frequent').fit_transform(data_train_cat),columns=data_train_cat.columns)
data_train_mean01=pd.DataFrame(SimpleImputer(strategy='mean').fit_transform(data_train_num),columns=data_train_num.columns)
```

```python
data_simple_imputed2=pd.concat([data_train_num01,data_train_cat01.drop('Loan Status',axis=1)],axis=1)
data_simple_imputed3=pd.concat([data_train_mean01,data_train_cat01.drop('Loan Status',axis=1)],axis=1)
```

```python
data_simple_imputed2.to_csv('Training data(median).csv',index=False)
data_simple_imputed3.to_csv('Training data(mean).csv',index=False)
```

```python
from sklearn.impute import KNNImputer
data_train_knn=pd.DataFrame(KNNImputer().fit_transform(data_train_num),columns=data_train_num.columns)
data_valid_knn=pd.DataFrame(KNNImputer().fit_transform(data_valid_num),columns=data_valid_num.columns)
```

```python
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer
data_train_iter=pd.DataFrame(IterativeImputer().fit_transform(X=data_train_num,y=data_train_num['Loan Status']),columns=data_train_num.columns)
#data_train_iter['Loan Status']=data_train_num['Loan Status']
data_valid_iter=pd.DataFrame(IterativeImputer().fit_transform(X=data_valid_num,y=data_train_num['Loan Status']),columns=data_valid_num.columns)
#data_valid_iter['Loan Status']=data_valid_num['Loan Status']
```

```python
data_iter_imputed_valid0=pd.concat([data_valid_iter,data_valid_cat01],axis=1)
data_iter_imputed=pd.concat([data_train_iter,data_train_cat01.drop('Loan Status',axis=1)],axis=1)
```
