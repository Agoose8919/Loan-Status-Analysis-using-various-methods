```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("Loans_Training Set.csv")
```

```python
data.isnull().sum()/len(data)
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.drop_duplicates(subset="Customer ID",inplace=True)
```

```python
#dropping column with more than 30%null value
#data.drop(['Months since last delinquent'],axis=1,inplace=True)
```

```python
data
```

```python
data['Loan Status']=data['Loan Status'].map({'Charged Off':1,'Fully Paid':0})
```

```python
data
```

```python
data["Maximum Open Credit"]= pd.to_numeric(data["Maximum Open Credit"],errors = 'coerce')
```

```python
data['Monthly Debt']=data['Monthly Debt'].replace({"\$":""},regex=True).astype(float)
```

```python
#Label encoding 
data["Years in current job"]= data["Years in current job"].replace({"< 1 year":0,"1 year":1,"2 years":2,"3 years":3,
                                                                      "4 years":4,"5 years":5,"6 years":6,"7 years":7,
                                                                      "8 years":8,"9 years":9,"10+ years":10})
#data['Years in current job'] = data['Years in current job'].str.replace(r'[^\d.]+', '',regex=True).astype('float')
```

```python
data
```

```python
data.drop(['Customer ID','Loan ID'],axis=1,inplace=True)
```

```python
data.isnull().sum()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in data.select_dtypes('number').columns:
    sns.distplot(data[i])
    plt.show()
```

```python
for i in data.select_dtypes('number').columns:
    sns.boxplot(data[i])
    plt.show()
```

```python
for i in data.select_dtypes('number').columns:
    sns.histplot(x=i,data=data)
    plt.show()
```

```python
for i in data.select_dtypes('object').columns:
    sns.countplot(data[i])
    plt.show()
```

```python
#Since credit score cannot be above 800,dropping rows with creditscore above 800
data['Credit Score']=np.where(data['Credit Score']<=800,data['Credit Score'],np.nan)
```

```python
#Google search shows that current limit for personal loan is 100k dollars,from any mainstream lenders. So removing current loan
#amounts above that from dataset.
data['Current Loan Amount']=np.where(data['Current Loan Amount']!=99999999,data['Current Loan Amount'],np.nan)
```

```python
max(data['Current Loan Amount'])
```

```python
len(data)
```

```python
#As we can see Debt Consolidation forms majority of purpose in this data. We will transform it to Debt Consolidation,not 
#Debt Consolidation.
set(data.Purpose)
```

```python
#data['Purpose']=np.where(data['Purpose']=='Debt Consolidation',1,0)#1 indicates purpose is debt consolidation,0 it isn't.
```

```python
data.Purpose.isnull().sum()
```

```python
#Labelling Term column,since it only has 2 distinct values,one hot encoding would result in same results
data['Term']=np.where(data['Term']=='Short Term',0,1)
```

```python
data['Home Ownership']=np.where(data['Home Ownership']=='HaveMortgage','Home Mortgage',data['Home Ownership'])
```

```python
sns.countplot(data['Home Ownership'])
```

```python
data['Current Loan Amount'].groupby(by=data['Loan Status']).mean()
```

```python
data['Maximum Open Credit']=np.where(data['Maximum Open Credit']<=95000,data['Maximum Open Credit'],np.nan)#Max credit limit in usa is 95k
```

```python
data.corr()
```

```python
len(data)
```

```python
#Checking distribution after cleaning
for i in data.select_dtypes('number'):
    sns.distplot(data[i])
    plt.show()
```

```python
data.to_csv('Loan data cleaned.csv',index=False)
```

```python
#fitting a model without removing any data
from catboost import CatBoostClassifier,Pool
```

```python
from sklearn.metrics import *
from sklearn.model_selection import train_test_split as tts
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
x=data.drop(['Loan Status','Purpose'],axis=1)
y=data['Loan Status']
```

```python
x_train,x_valid,y_train,y_valid=tts(x,y,test_size=0.2,stratify=y,random_state=365)
```

```python
pool_train=Pool(x_train,y_train,cat_features=['Term','Home Ownership'])
pool_valid=Pool(x_valid,y_valid,cat_features=['Term','Home Ownership'])
```

```python
help(CatBoostClassifier)
```

```python
cb=CatBoostClassifier(auto_class_weights='Balanced')
```

```python
cb.fit(pool_train)
```

```python
#Using metrics to test the model
pd.DataFrame(data=[accuracy_score(y_train,cb.predict(pool_train)), recall_score(y_train,cb.predict(pool_train)),
                   precision_score(y_train,cb.predict(pool_train)), roc_auc_score(y_train,cb.predict(pool_train)),
                   f1_score(y_train,cb.predict(pool_train))], 
             index=["accuracy", "recall", "precision", "roc_auc_score","f1_score"])
```

```python
print(classification_report(y_train,cb.predict(pool_train)))
```

```python
#Using metrics to test the model
pd.DataFrame(data=[accuracy_score(y_valid,cb.predict(pool_valid)), recall_score(y_valid,cb.predict(pool_valid)),
                   precision_score(y_valid,cb.predict(pool_valid)), roc_auc_score(y_valid,cb.predict(pool_valid)),
                   f1_score(y_valid,cb.predict(pool_valid))], 
             index=["accuracy", "recall", "precision", "roc_auc_score","f1_score"])
```

```python
print(classification_report(y_valid,cb.predict(pool_valid)))
```
