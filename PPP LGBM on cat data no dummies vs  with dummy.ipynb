Baseline LGBM Model

```python
lgbm=LGBMClassifier(is_unbalance=True,n_jobs=4)
```

```python
x=data.drop(['Loan Status'],axis=1)
y=data['Loan Status']
```

```python
x_train,x_valid,y_train,y_valid=tts(x,y,test_size=0.2,stratify=y,random_state=365)
```

```python
lgbm.fit(x_train,y_train)
```

```python
lgbm.predict_proba(x_valid)[:,1]
```

```python
from sklearn.metrics import *
print(classification_report(y_valid,lgbm.predict(x_valid)))
```

```python
#Using metrics to test the model
pd.DataFrame(data=[accuracy_score(y_valid,lgbm.predict(x_valid)), recall_score(y_valid,lgbm.predict(x_valid)),
                   precision_score(y_valid,lgbm.predict(x_valid)), roc_auc_score(y_valid,lgbm.predict_proba(x_valid)[:,1]),
                   f1_score(y_valid,lgbm.predict(x_valid),average='macro')], 
             index=["accuracy", "recall", "precision", "roc_auc_score","f1_score"])
```
